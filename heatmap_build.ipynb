```python
def create_heatmap_kml(fileName, mapType):
    for i in range(start, end):
        index = i-start
        point = mapPoints[i]
        bounding_box = boundingBoxes[i]
        description = ("<![CDATA[" + "\n"+ 
                       "<p>PointId: " + str(i+1) + "</p>" +"\n" + 
                       "<p>Latitude: " + str(point[0]) + "</p>" + "\n" +
                       "<p>Longitude: " + str(point[1]) + "</p>" +  "\n" +
                       "<p>Travel Time: " + str(hospital['duration'][index]) +" min" + "</p>" + "\n" +
                       "<p>Distance: " + str(hospital['distance'][index])+ " mi" + "</p>" + "\n" + 
                       "]]>")
        print(description)
        _name = hospital_map[hospital['hosp_id'][index]][0] + ", " + hospital_map[hospital['hosp_id'][index]][1]
        pol = kml.newpolygon(name=_name,
                         outerboundaryis=[(bounding_box[1], bounding_box[0]),
                                          (bounding_box[1], bounding_box[2]),
                                          (bounding_box[3], bounding_box[2]),
                                          (bounding_box[3], bounding_box[0]),
                                          (bounding_box[1], bounding_box[0])])
        value = 0
        color = ''
        if mapType == 'distance':
            value = hospital['distance'][index]
            if value < 5:
                color = '7200ff00'
            elif value >= 5 and value <10:
                color = '7200ffe8'
            elif value >= 10 and value<15:
                color = '7200ffe3'
            elif value >= 15 and value<20:
                color = '7200ffff'
            elif value >= 20 and value<25:
                color = '7200d1ff'
            elif value >= 25 and value<30:
                color = '720094ff'
            elif value >= 30:
                color = '720000ff'
        else:
            value = hospital['duration'][index]
            if value < 5:
                color = '7200ff00'
            elif value >= 5 and value<10:
                color = '7200ffe8'
            elif value >= 10 and value<15:
                color = '7200ffe3'
            elif value >= 15 and value<25:
                color = '7200ffff'
            elif value >= 25 and value<35:
                color = '7200d1ff'
            elif value >= 35 and value<45:
                color = '720094ff'
            elif value >= 45:
                color = '720000ff'
            
        pol.description = description
        pol.style.polystyle.color = color
        pol.style.polystyle.outline = 0
        pol.style.polystyle.fill = 1
    
    kml.save(fileName)
```
